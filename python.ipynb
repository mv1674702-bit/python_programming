```python
def list_count(num):
    count=0
    for i in num:
        if i == 4:
             count+=1
    return count
    
print(list_count([3,4,6,4,2,7,5,4]))
```

```python
string=input("enter the string:")
n=int(input("enter the number:"))
if len(string)>2:
    print(string[0:2]*n)
else:
    print(string*n)
```
